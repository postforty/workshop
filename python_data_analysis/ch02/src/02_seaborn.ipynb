```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np # 난수 생성을 위해 numpy 추가

# 샘플 데이터셋 생성 (tips 데이터셋 모방)
np.random.seed(42) # 재현성을 위해 시드 설정
data_size = 50

# total_bill: 10부터 50 사이의 값
total_bill = 10 + 40 * np.random.rand(data_size) 
# tip: total_bill의 약 10%~20% 범위
tip = total_bill * (0.1 + 0.1 * np.random.rand(data_size)) 
# time: Lunch(0), Dinner(1)
time_categories = ['Lunch'] * (data_size // 2) + ['Dinner'] * (data_size - data_size // 2)

tips = pd.DataFrame({
    'total_bill': total_bill,
    'tip': tip,
    'time': time_categories
})

# 범주형 데이터 타입을 명시적으로 지정 (Seaborn의 hue 인수를 위해)
tips['time'] = tips['time'].astype('category')

# 산점도 생성
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np # 난수 생성을 위해 numpy 추가

# 샘플 데이터셋 생성
np.random.seed(42) # 재현성을 위해 시드 설정
data_size = 50

# 숫자형 컬럼 생성 (tips 데이터셋 모방)
total_bill = 10 + 40 * np.random.rand(data_size) 
tip = total_bill * (0.1 + 0.1 * np.random.rand(data_size)) 
size = np.random.randint(1, 7, size=data_size) # 일행 인원

# 범주형 컬럼 생성 (tips 데이터셋 모방)
days = np.random.choice(['Thur', 'Fri', 'Sat', 'Sun'], size=data_size)
sex = np.random.choice(['Male', 'Female'], size=data_size)
smoker = np.random.choice(['Yes', 'No'], size=data_size)
time = np.random.choice(['Lunch', 'Dinner'], size=data_size)

tips = pd.DataFrame({
    'total_bill': total_bill,
    'tip': tip,
    'sex': sex,
    'smoker': smoker,
    'day': days,
    'time': time,
    'size': size
})

# 범주형 데이터 타입을 명시적으로 지정
tips['day'] = tips['day'].astype('category')
tips['sex'] = tips['sex'].astype('category')
tips['time'] = tips['time'].astype('category')

# -------------------------------------------------------------

# 히트맵 생성
# 숫자형 컬럼만 선택하여 상관관계 행렬 계산
corr = tips.select_dtypes(include=['number']).corr() 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# -------------------------------------------------------------

# 박스플롯 생성
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title('Total Bill by Day and Sex')
plt.show()
```
